# **2. Data Wrangling**
  



### Data Cleaning:
- Check for missing values and count them per column
- Drop all rows where the title or type is missing
- Convert the date_added column to datetime format

```python
# Check for missing values and count them per column

missing_values = df.isnull().sum()
print(f"Count for Missing Values \n{missing_values}\n per column")

# Drop all rows where the title or type is missing
df_clean = df.copy()

df_clean.dropna(subset=['title', 'type'], inplace=True)
print("\n\nDropping all Rows where the title or type is missing\n")
df_clean

# Convert the `date_added` column to datetime format
df_clean['date_added'] = pd.to_datetime(df_clean['date_added'], format='mixed')   # dates in string format like `2024-07-01', 'July 3, 2025`
df_clean.info()

```

```python
# Handle missing values with Descriptive Statistics

df_clean['director'].fillna('unknown', inplace=True)

df_clean['country'].mode()  # most frequent value
df_clean['country'].fillna('United States', inplace=True)

df_clean['cast'].fillna('not specified', inplace=True)

df_clean['rating'].mode()  # most frequent value
df_clean['rating'].fillna('TV-MA', inplace=True)

df_clean['duration'].fillna('not specified', inplace=True)

df_clean
```

```python
df_clean.info()
```

### Filtering and indexing:
- Show only movies released after 2015
- Filter all shows from the country India
- Select only Movie type entries with duration over 90 minutes

```python
# Show only movies released after 2015
movies_after_2015 = df_clean[(df_clean['type'] == 'Movie') & (df_clean['release_year'] > 2015)]
movies_after_2015
```

```python
# Filter all shows from the country India
india_shows = df_clean[df_clean['country'] == 'India']
india_shows
```

```python
# Select only Movie type entries with duration over 90 minutes

# Filter only movies
movies = df_clean[df_clean['type'] == 'Movie'].copy()

# Extract the numeric part of the duration, fill NaN values with 0, and convert to integer
movies['duration_numeric'] = movies['duration'].str.extract('(\d+)').fillna(0).astype(int)

# Filter movies with duration over 90 minutes
movies_over_90_minutes = movies[movies['duration_numeric'] > 90]
movies_over_90_minutes
```

### Column operations:
- Create a new column called `release_decade` based on the release_year
- Display unique values in type, rating, and country

```python
# Create a new column called release_decade based on the release_year
df_clean['release_decade'] = (df_clean['release_year'] // 10) * 10
df_clean
```

```python
# Display unique values in type, rating, and country
unique_types = df_clean['type'].unique()   # Unique value in `Type`

unique_ratings = df_clean['rating'].unique()    # Unique value in `Rating`

unique_countries = df_clean['country'].unique()    # Unique value in `Country`

print(f"Unique Types: \n{unique_types}\n")
print(f"Unique Ratings: \n{unique_ratings}\n")
print(f"Unique Countries: \n{unique_countries}\n")
```

## Aggregations:
- Group by type and count the number of titles in each category
- Group by country and find the top 5 countries with the most titles
- Group by rating and find average release year for each rating type

```python
# Group by type and count the number of titles in each category
type_counts = df_clean.groupby('type')['title'].count()
print(type_counts)
```

```python
# Group by country and find the top 5 countries with the most titles
country_counts = df_clean.groupby('country')['title'].count()
top_5_countries = country_counts.nlargest(5)
print(top_5_countries)
```

```python
# Group by rating and find average release year for each rating type
rating_release_year = df_clean.groupby('rating')['release_year'].mean()
print(rating_release_year)
```

## Multiple aggregations:

```python
# Group by release_year and show
df_clean.groupby('release_year')['type'].value_counts()
```

```python
# Total number of titles
total_titles = df_clean.groupby('release_year')['title'].count()
print(f"Total_titles: \n{total_titles}")

# Count of Movies and TV Shows separately
type_counts = df_clean.groupby(['release_year', 'type'])['title'].count().unstack(fill_value=0)
print(f"\nType Counts: \n{type_counts}")
```

## Advanced:
- Create a new column `is_recent` True if `release_year` > 2015
- Group by country and count how many recent shows are there per country

```python
#  Create a new column is_recent True if release_year > 2015
df_clean['is_recent'] = df_clean['release_year'] > 2015
df_clean
```

```python
# Group by country and count how many recent shows are there per country
recent_shows_per_country = df_clean[df_clean['is_recent']].groupby('country')['title'].count()
print(f"Recent Shows: {recent_shows_per_country}")
```
